### Decision sobre codigos y valores ambiguos


Los valores `P1`, `P2`, `P3` y `P4` se interpretan como una escala de prioridad donde `P1` es la mas critica. En cambio, los valores numericos aislados (`1`, `2`, `3`, `4`) son ambiguos: podrian representar una escala ascendente o descendente. Se buscó observar los comentarios manualmente , se nota una tendencia a fallas funcionales de dispositivos en prioridad 1, pero tambien se encuentran este tipo de obervaciones para tickets con prioridad 4 


Por esa razon, los numeros aislados no se mapean automaticamente. Se conservan como `Unknown` en `clean_ticket_priority` para evitar introducir una suposicion no verificable. Mas adelante, el LLM podra generar una columna separada (`ai_suggested_priority`) basada en el contenido del ticket.

```python
priority_mapping = {
    "critical": "Critical",
    "urgent": "Critical",
    "p1": "Critical",

    "high": "High",
    "alta": "High",
    "p2": "High",

    "medium": "Medium",
    "med": "Medium",
    "p3": "Medium",

    "low": "Low",
    "baja": "Low",
    "p4": "Low",
}
df["clean_ticket_priority"] = (
    df["Ticket Priority"]
    .map(priority_mapping)
    .fillna("Unknown")
)
```

Con el mapeo anterior se crea `clean_ticket_priority`, una version limpia de prioridad que conserva como `Unknown` los nulos y los valores ambiguos. Esta columna sera la base para KPIs y filtros del dashboard.

```python
df[["First Response Time","Time to Resolution"]]
```

Se usa un formato explicito (`%Y-%m-%d %H:%M:%S`) porque estas dos columnas parecen seguir una estructura consistente. `errors="coerce"` permite convertir valores invalidos o vacios en `NaT`, sin detener la ejecucion.

## 10. Limpieza de `Date of Purchase`

Para manejar fechas en espanol, se normalizan los nombres de meses a ingles y se elimina el conector `de`. Esto convierte valores como `19 de diciembre 2020` en `19 december 2020`.

```python
spanish_months = {
    "enero": "january",
    "febrero": "february",
    "marzo": "march",
    "abril": "april",
    "mayo": "may",
    "junio": "june",
    "julio": "july",
    "agosto": "august",
    "septiembre": "september",
    "setiembre": "september",
    "octubre": "october",
    "noviembre": "november",
    "diciembre": "december",
}
def normalize_spanish_date_text(value):
    if pd.isna(value):
        return value

    text = str(value).strip().lower()

    for spanish_month, english_month in spanish_months.items():
        text = text.replace(spanish_month, english_month)

    text = text.replace(" de ", " ")

    return text
```

```python
df["purchase_date_text_normalized"] = df["Date of Purchase"].apply(
    normalize_spanish_date_text
)
```

```python
df[["Date of Purchase", "purchase_date_text_normalized"]].head(20)
```

```python
df["clean_date_of_purchase"] = pd.to_datetime(
    df["purchase_date_text_normalized"],
    errors="coerce"
)
```

La funcion `parse_purchase_date` intenta varios formatos explicitamente. Esto evita depender de una inferencia automatica de pandas, que no funciona bien cuando la misma columna mezcla formatos.

```python
def parse_purchase_date(value):
    if pd.isna(value):
        return pd.NaT

    normalized = normalize_spanish_date_text(value)

    formats = [
        "%Y-%m-%d",      # 2021-07-19
        "%m-%d-%Y",      # 10-23-2020
        "%d/%m/%Y",      # 28/04/2021
        "%d %B %Y",      # 19 december 2020
    ]

    for date_format in formats:
        parsed = pd.to_datetime(
            normalized,
            format=date_format,
            errors="coerce"
        )

        if pd.notna(parsed):
            return parsed

    return pd.NaT
```

```python
df["clean_date_of_purchase"] = df["Date of Purchase"].apply(parse_purchase_date)
```

```python
df[["Date of Purchase", "purchase_date_text_normalized", "clean_date_of_purchase"]].head(20)
```

La validacion final confirma que `clean_date_of_purchase` conserva solo los 31 nulos originales. No quedaron fechas no nulas sin convertir.

## 14. Enriquecimiento mock de IA

```python
df.head()
```

```python
def assign_responsible_team(ticket_type):
    mapping = {
        "Technical Issue": "Technical Support",
        "Billing Inquiry": "Billing",
        "Refund Request": "Billing",
        "Cancellation Request": "Customer Success",
        "Product Inquiry": "Customer Support",
    }
    return mapping.get(ticket_type, "General Support")
def detect_sentiment(text):
    text = str(text).lower()

    urgent_words = ["urgent", "critical", "immediately", "as soon as"]
    negative_words = ["issue", "problem", "error", "unable", "can't", "bug", "failed"]

    if any(word in text for word in urgent_words):
        return "Urgent"
    if any(word in text for word in negative_words):
        return "Frustrated"

    return "Neutral"
def suggest_priority(row):
    text = f"{row['Ticket Subject']} {row['Ticket Description']}".lower()

    if row["clean_ticket_priority"] != "Unknown":
        return row["clean_ticket_priority"]

    if any(word in text for word in ["urgent", "critical", "unable", "not working", "failed"]):
        return "High"

    if any(word in text for word in ["refund", "payment", "billing", "cancel"]):
        return "Medium"

    return "Low"
def create_mock_summary(row):
    return (
        f"Cliente reporta {row['clean_ticket_type'].lower()} "
        f"relacionado con {row['product_text']}. "
        f"Asunto: {row['Ticket Subject']}."
    )
```

```python
df["ai_category"] = df["clean_ticket_type"]

df["ai_suggested_priority"] = df.apply(suggest_priority, axis=1)

df["ai_summary"] = df.apply(create_mock_summary, axis=1)

df["ai_sentiment"] = df["Ticket Description"].apply(detect_sentiment)

df["ai_responsible_team"] = df["clean_ticket_type"].apply(assign_responsible_team)
```

```python
df["ai_sentiment"].value_counts()
df["ai_responsible_team"].value_counts()
```
